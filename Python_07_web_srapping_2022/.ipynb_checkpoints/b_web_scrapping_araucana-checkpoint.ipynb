Podemos entender el scraping como una serie de técnicas para obtener datos desde un archivo de la web. Puede ser culquier tipo de página web html. Para esto se va a una URL (dirección web) y se explora su estructura HTML, tomando los elementos que se necesiten. 

Para hacer scraping exsiten muchas opciones, en Python existe una librería llamada Beautiful Soup que permite explorar la estructura de un HTML y obtener etiquetas HTML desde esta. 

En este caso obtendremos el texto del CANTO I de "La Araucana" de Ercilla, disponible en wikipedia. 

Ver más en: https://issuu.com/ricardov_net/docs/datos_obtencion_scraping_2022


```python
## Importar las librerias a utilizar  
import pandas as pd
import time

from bs4 import BeautifulSoup
import requests
```

```python
## URL de texto en la web 
ruta = "https://es.wikisource.org/wiki/La_araucana_primera_parte:_008"

## solicitar la web
req = requests.get(ruta )
soup =  BeautifulSoup(req.text, "lxml")

## Mostrar el texto que obtiene, es un HTML con sus etiquetas <title>, <h1>, <p>, etc.
soup
```

## A- Capturar Parrafo Inicial

```python
## A1– captura tit
titpagina = soup.find("b")
#print("TIT PAG: ", titpagina)


## A2– captura parrafo inicial primera linea
parr_inicio = soup.find("i")
print("PARRAFO INICIO: ", parr_inicio)

```

```python
## A3– Captura parrafo inicial todo
## Elemento a buscar de tipo i (italicas)
parr_inicio = soup.find_all("i")
print("PARRAFO TODO: ", parr_inicio)


## largo de esta lista
len(parr_inicio)

```

```python
## A4– Captura parrafo inicial y mostrar sus elementos

## Elemento a buscar de tipo i
parr_inicio = soup.find_all("i")

## Recorrer 
for i in range( len(parr_inicio)  ): 
    print("ELEMENTO: ", parr_inicio[i] )

    
```

## B- Captura Texto, elementos de un tipo

```python
## B1– Captura varios parrafos iniciales

## Elemento a buscar de tipo p, parrafo
lineas = soup.find_all("p")


## Mostrar que hay en lineas
lineas

```

```python
## B2– Recorrer lo que hay en lineas

for i in range(len(lineas)) :    
    print( '\n', i, '- - - - - - -\n',lineas[i] )
    pass
    
## Problema, tenemos que limpiar una serie de elementos, los tags 
```

```python
## B3– Rescatar solo el texto

result = soup.find_all("p")
texto  = []


## Recorreomos la cantidad de elementos
for i in range( len(result) ):
    t = result[i].text # en t guarda el texto del elemento
    texto.append( t ) # lo colocamos en la lista
    #print(i,'- - - - \n', t)
    
## En este caso, el texto de interes esta en un elemento de la lista
texto
```

```python
## B4– Tomeamos el elemento 2 y lo vamos a separar en sus saltos de linea \n
texto_a_separar = texto[2]
texto_a_separar
```

```python
## B5– Separando

## Este texto lo podemos separar con el metodo split
texto_dividido =  texto_a_separar.split('\n')
texto_dividido
```

## C- Guardar archivo

```python
## C1– Lo pasamos a un df paraluego guardarlo como CSV

df_canto01 = pd.DataFrame( { 'titulo_canto':"Canto 1", 'texto':texto_dividido } )

df_canto01
```

```python
## Y lo guardamos en un archivo csv

## Ojo, el archivo se llama canto01.csv y se guarda en la carpeta txt_araucana
## Si no existe dicha carpeta debe crearla
df_canto01.to_csv("txt_araucana/canto01.csv")
```
