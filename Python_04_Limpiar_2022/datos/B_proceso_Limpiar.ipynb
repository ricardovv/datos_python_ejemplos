# 5- VISUALIZAR BARRAS

```python
# Imprtar libreria visualizacion
import matplotlib.pyplot as plt
```

```python
# Mostrar como barras
x1 = df['I_x']
y1 = df['I_y']
plt.bar(x=x1, height=y1)
```

```python
# Tamanho
fig=plt.figure(figsize=(5, 2))
#Titulo 
plt.title("El Famoso Cuarteto de Anscombe")
## Eje X e Y etiquetas. 
plt.xlabel("Este es el Eje X")
plt.ylabel("Este es el Eje Y")
# Grafico de Barras
plt.bar(x=x1 ,height=y1)
plt.savefig('A_datos/a_cuarteto_barra.jpg')
plt.show()
```

# 6- VISUALIZAR SCATTERPLOT

```python
# Tamanho
fig=plt.figure(figsize=(6, 4))
#Titulo 
plt.title("El Famoso Cuarteto de Anscombe")
## Eje X e Y etiquetas. 
plt.xlabel("Este es el Eje X")
plt.ylabel("Este es el Eje Y")
# Grafico de Barras
plt.scatter(x=x1, y=y1, color='red')
plt.savefig('A_datos/a_cuarteto_scatter.jpg')
plt.show()

#ax.scatter(grades_range, girls_grades, color='r')

```
