# 3- LIMPIAR
###### En la etapa de limpieza por lo general se formatean los datos, se unifican valores con puntos o comas, se colocan textos en baja, se detecta informacion incompleta que en ciertos casos se completa o elimina. Dependiendo de los datos, del problema o del escenario, pueden requerirse otras operaciones. Revisaremos algunas. 
###### Se recomienda que el resultado de la limpieza se guarde en un archivo para ser usado en etapas siguientes. 

```python
# botar lineas sin datos, incompletos o con muchos campos vacios.
# el resultado lo guardaremos en el dataframe dflimpiando.
# ojo, la fila 5 desaparece, pero los index no estan actualizados...
dflimpiando = df.drop( [5, 11] ) # drop bota las lineas seleccionadas 
dflimpiando
```

```python
# identificar valores unicos en una columna para ver cuales estan mal escritos. 
# fijarse cuantas maneras esta escrito un elemento, se deben unificar. 
dflimpiando['item'].unique()  
```

```python
# en dicha columna, se reemplazan valores malos por los correctos.
dflimpiando = dflimpiando.replace({'item':{'LAPI':'lÃ¡piz',
                                           'LAPICS':'lÃ¡piz',
                                           'lapiz':'lÃ¡piz',
                                           'LIBRO':'libro'}
                                  })
# mostrar dataframe dflimpiando.
dflimpiando
```

```python
# poner valores de una columna en bajas.  
dflimpiando['nombre'] = dflimpiando['nombre'].str.lower() 
# mostrar dataframe dflimpiando.
dflimpiando
```

```python
# fechas estan en diferentes formatos, se debe unificar. 
dflimpiando['fecha'] = pd.to_datetime(dflimpiando.fecha)
```

```python
# mostramos el dflimpiando para ver como va quedando.
# ahora parece algo mas ordenado 
dflimpiando
```

# 4- NUEVA COLUMNA CON VALORES 

```python
# crear nueva columna en base a otras. 
# valor total desde valor unitario x cantidad.
dflimpiando['total'] = dflimpiando['cantidad']* dflimpiando['costounidad']
dflimpiando
```

# 5- GUARDAR ARCHIVO LIMPIO
###### Si estamos ok con nuestro archivo limpio, lo guardamos para trabajar el analisis con este. 

```python
# guardar datos limpios para usarlos en los procesos siguientes. 
dflimpiando.to_csv('guardar/a_comprasutiles_limpio.csv')
```
