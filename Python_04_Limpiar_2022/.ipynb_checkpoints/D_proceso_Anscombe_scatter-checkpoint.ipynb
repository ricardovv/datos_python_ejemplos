# D- CUARTETO DE ANSCOMBE - SCATTERPLOT 
#### Ejercicio simple para reconocer el proceso

# 1- IMPORTAR LIBRERIAS Y CARGAR DATOS

```python
# importar librerias. Pandas es una libreria de python para trabajar con datos. 
# se lee "importar la libreria PANDAS con el nombre pd".
import pandas as pd
# importar libreria visualizacion.
import matplotlib.pyplot as plt
```

```python
# cargar datos. pd.read_csv para leer archivos .csv
# hay comandos para leer json, excel, entre otros.   
df = pd.read_csv('A_datos/cuarteto_anscombe.csv') ###  la ruta y nombre del archivo.

# mostrar el dataframe donde se guardaron los datos
# df 
```

```python
# creamos las variables de las seriede de datops.
x1 = df['I_x'] # en x1 guardaremos los datos de la columna'I_x'.
y1 = df['I_y'] # en y1 guardaremos los datos de la columna'I_y'.
```

# 2- VISUALIZAR SCATTERPLOT: SOLO UNA SERIE
###### en este scaterpot graficaremos solo un conjunto de datos, I. 

```python
# ahora probaremos con grafico de puntos.
fig=plt.figure(figsize=(6, 4)) # tamano del grafico. 

plt.title("El Famoso Cuarteto de Anscombe, solo una serie X1 e Y1") # titulo. 
plt.xlabel("Este es el Eje X") # texto eje X. 
plt.ylabel("Este es el Eje Y") # texto eje Y. 

plt.scatter(x=x1, y=y1, color='red') # grafico de puntos.

# guarda el grafico en la ruta seleccionada.
plt.savefig('A_datos/a_cuarteto_scatter_simple.jpg')
# esta linea muestra el grafico.
plt.show() 
```

# 3- VISUALIZAR SCATTERPLOT DOS SERIES
###### en este scaterpot graficaremos dos conjuntos de datos, I y II.
###### como ejercicio puede incorporar las  series III y IV y adaptar colores y titulos.

```python
### Datos - - - - - - - - - - - - 
# guardamos datos en dos variables, grupo I en g1, y grupo II en g2.
g1 = (df['I_x'], df['I_y'])
g2 = (df['II_x'], df['II_y'])

gdatos = (g1, g2) # guardamos esos grupos en la variable datos.
gnombres = ('El grupo 1', 'El grupo 2') # guardamos esos grupos en la variable datos.
gcolores = ('red', 'blue')


### Crear grafico - - - - - - - - - - - - 
fig=plt.figure(figsize=(10, 6), facecolor='lightslategray') # tamano del grafico. 
fig.patch.set_facecolor('green') # color del margen.
fig.patch.set_alpha(0.2) # transparencia del margen.

ax = fig.add_subplot(1, 1, 1) # agregar sub graficos en el mismo espacio.
ax.patch.set_color('blue') # transparencia del grafico. 
ax.patch.set_alpha(0.9) # transparencia del grafico. 

plt.title("El Famoso Cuarteto de Anscombe") # titulo. 
plt.xlabel("Este es el Eje X") # texto eje X. 
plt.ylabel("Este es el Eje Y") # texto eje Y. 
plt.grid( color='grey', alpha=0.3) # grilla.

# se recorre la info de gdatos, gcolores, gnombres y se guardan en dat, col, gru. 
for dat, col, gru in zip(gdatos, gcolores, gnombres):
    x, y = dat
    ax.scatter(x, y, alpha=0.6, c=col, edgecolors='none', s=80, label=gru, marker='o') # puntos.
    plt.legend(gnombres) # leyendas.

    
### Mostrar y guardar grafico - - - - - - - - - - - -     
# guarda el grafico en la ruta seleccionada.
plt.savefig('A_datos/a_cuarteto_scatter_series.jpg')
# esta linea muestra el grafico.
plt.show() 
```
